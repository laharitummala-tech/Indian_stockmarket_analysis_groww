```python
def clean_and_combine(file_date):
    cleaned_dfs = []
    for file,date in file_date:
        df = pd.read_csv(file)
        df["Date"] = pd.to_datetime(date)
        df = df.drop_duplicates()
        cleaned_dfs.append(df)
    combined = pd.concat(cleaned_dfs,ignore_index=True)
    combined = combined.drop_duplicates().reset_index(drop = True)
    return combined
```

```python
files = [
    ("day_1_groww.csv", "2026-02-17"),
    ("day_2_groww.csv", "2026-02-18"),
    ("day_3_groww.csv", "2026-02-19"),
    ("day_4_groww.csv", "2026-02-20"),
    ("day_5_groww.csv", "2026-02-23")
]
final_df = clean_and_combine(files)
```
